```python
#convert categorical to numerical using dummies variable
#d1=pd.get_dummies(car.sell_type)
d1 = pd.get_dummies(car.Fuel_Type)
d2=pd.get_dummies(car.Transmission)
d3=pd.get_dummies(car.Seller_Type)
cars=pd.concat([car,d1,d2,d3,],axis='columns')
```

```python
cars.drop(['Car_Name','Fuel_Type','Seller_Type','Transmission','Individual','Owner','Dealer','Automatic','CNG'],axis=1,inplace=True)
```

```python
x=cars.drop('Selling_Price',axis=1)
y=cars.iloc[:,-6].values
```

```python
#cross validation of the data
from sklearn.model_selection  import train_test_split
from sklearn.linear_model import LinearRegression
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
lr=LinearRegression()
print(lr.fit(X_train,y_train))
print(lr.fit(X_test,y_test))
print(lr.score(X_train,y_train))
print(lr.score(X_test,y_test))
lrpred=lr.predict(X_test)
print('linear regresiion prediction{}'.format(lrpred))
```

```python
plt.scatter(y_test,lrpred,color='g')
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
resudual=y_test-lrpred
print(resudual)
sns.distplot(resudual)
```

```python
df = pd.DataFrame({'Actual':y_test.flatten(), 'Predicted':lrpred.flatten()})
print(df)
```

```python
from sklearn import metrics
print(metrics.mean_absolute_error(y_test,lrpred))
print(metrics.mean_squared_error(y_test,lrpred))
print('rmse',np.sqrt(metrics.mean_squared_error(y_test,lrpred)))
print('r2score',metrics.r2_score(y_test,lrpred))
```

```python
#used optimiztion to get the better orediction
from sklearn.linear_model import Ridge 
rd=Ridge(alpha=1)
rd.fit(X_train,y_train)
rd.fit(X_test,y_test)
rd.score(X_train,y_train)
rd.score(X_test,y_test)
rdpred=rd.predict(X_test)
print(rdpred)
plt.scatter(y_test,rdpred)
df1=pd.DataFrame({'Actual':y_test.flatten(), 'Predicted':rdpred.flatten()})
print(df)
print(metrics.mean_absolute_error(y_test,rdpred))
print(metrics.mean_squared_error(y_test,rdpred))
print('rmse',np.sqrt(metrics.mean_squared_error(y_test,rdpred)))
print('r2score',metrics.r2_score(y_test,rdpred))
```
